```python
# Display dataset structure
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(path)
df = pd.DataFrame(data)
print("Dataset Overview:")
print(df)


#Line Graph (using Seaborn)
plt.plot(figsize=(8, 6))
sns.lineplot(data=data, x='Temperature', y='Humidity', label='Temp vs Humidity')
sns.lineplot(data=data, x='Temperature', y='Pressure', label='Temp vs Pressure')
plt.title("Line Graph: Temperature vs Humidity/Pressure")
plt.xlabel("Temperature")
plt.ylabel("Values")
plt.show()


#Bar plot(Using seaborn)
sns.barplot(data=data, x='Temperature', y='Humidity', label='Temp vs Humidity')
sns.barplot(data=data, x='Temperature', y='Pressure', label='Temp vs Pressure ')
plt.xlabel('Temparature', fontsize=12, fontweight='bold')
plt.ylabel('Value', fontsize=12, fontweight='bold')
plt.show()

#Pie Chart
labels=['Temperature','Pressure','Humidity']
sizes=[data['Temperature'].mean(),data['Pressure'].mean(),data['Humidity'].mean()]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie Chart')
plt.show()

#Pie chart using Bokeh
!pip install bokeh
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource, LabelSet
from math import pi

output_notebook()
# calculating the percentages and angles for the pie chart
total = sum(sizes)
percentages = [(s / total) * 100 for s in sizes]
angles = [(p / 100) * 2 * pi for p in percentages]

# Creating a ColumnDataSource for the pie chart
source = ColumnDataSource(data=dict(
    labels=labels,
    sizes=sizes,
    percentages=percentages,
    angles=angles,
    starts=[0, angles[0], angles[0] + angles[1]],
    ends=[angles[0], angles[0] + angles[1], angles[0] + angles[1] + angles[2]]
))

# Creating the figure
p = figure(title="Simple Bokeh Plot", x_axis_label="Temparature", y_axis_label="Humidity")
p.wedge(x=0, y=0, radius=0.7, source=source,
        start_angle='starts', end_angle='ends', line_color="white",
         legend_field='labels')
labels = LabelSet(x=0, y=0, text='percentages', angle='angles',
                  text_font_size="8pt", text_align='center', source=source)
p.add_layout(labels)

hover = HoverTool(tooltips=[("Temperature", "@x"), ("Humidity", "@y")])
p.add_tools(hover)

p.legend.location = "top_left"

show(p)





#Using Bokeh
!pip install bokeh
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
output_notebook()
p = figure(title="Simple Bokeh Plot", x_axis_label="Temparature", y_axis_label="Humidity")
p.line(x=data['Temperature'], y=data['Humidity'], legend_label="Humidity",color="green", line_width=2)
p.line(x=data['Temperature'], y=data['Pressure'], legend_label="Pressure", line_width=2)
p.line(x=data['Temperature'], y=data['Humidity'], legend_label="Humidity",color="green", line_width=2)
p.line(x=data['Temperature'], y=data['Pressure'], legend_label="Pressure", line_width=2)
p.vbar(x=data['Temperature'], top=data['Humidity'], legend_label="Humidity",color="blue", width=0.5)
p.vbar(x=data['Temperature'], top=data['Pressure'], legend_label="Pressure",color="green", width=0.5)
p.legend.location = "top_left"
hover = HoverTool(tooltips=[("Temperature", "@x"), ("Humidity", "@y")])
p.add_tools(hover)
p.legend.location = "top_left"
hover = HoverTool(tooltips=[("Temperature", "@x"), ("Humidity", "@y")])
p.add_tools(hover)


show(p)








```
